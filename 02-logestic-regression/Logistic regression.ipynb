```python
#RFE : RECURSIVE FEATURE ELIMINATION
#its a feature selection technique.
#it trains the mode
#ranks the feature by importance
#it repeat the process recursively
#stops when the desired no of features are selected
#its going to mark the unimportant columns as False
#important columns are true
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
rfe = RFE(model)
```

```python
rfe.fit(x,y)
```

```python
rfe.support_
```

```python
selected_indexes = []
for i, data in enumerate(rfe.support_):
    if data == True:
        selected_indexes.append(i)
        
```

```python
selected_indexes
```

```python
li = ['Feature1','Feature2','Feature3','Feature4']
for i, x in enumerate(li):
    if x=='Feature3':
        print(i)
```

```python
dummy = df.iloc[:,selected_indexes]
dummy
```

```python
x = dummy
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
lr = LogisticRegression(max_iter=1000)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_pred,y_test)*100
```

```python
confusion_matrix(y_pred,y_test)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_pred,y_test),annot=True,cmap='coolwarm',fmt='d')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.show()
```

```python
x_test
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_probs = lr.predict_proba(x_test)[:,1]

fpr , tpr, threshold = roc_curve(y_test, y_probs)
auc = roc_auc_score(y_test,y_probs)

plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'ROC Curve(Auc={auc:.2f})')
plt.plot([0,1],[0,1],'k--',label='Random Classifer')
plt.xlabel('FalsePositive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.title('ROC CURVE')
plt.show()
```
