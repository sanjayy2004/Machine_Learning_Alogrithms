```python
for col in df.columns:
    if df[col].dtype == object:
        df[col].fillna(df[col].mode()[0],inplace=True)
    else:
        df[col].fillna(df[col].mean(),inplace=True)
        
```

```python
df.isnull().sum()
```

```python
#outliers
```

```python
for x in df.columns:
    if df[x].dtype != object:
        plt.boxplot(df[x])
        plt.xlabel(x)
        plt.ylabel('count')
        plt.show()
        
```

```python
for x in df.columns:
    if df[x].dtype != object or x == 'charges':
        Q1 = df[x].quantile(0.25)
        Q3 = df[x].quantile(0.75)
        IQR = Q3-Q1
        ll = Q1 - 1.5*IQR
        up = Q3 + 1.5*IQR
        df = df[(df[x]>=ll) & (df[x]<=up)]

```

```python
#VIF : variation inflation factor
#VIF helps us to detect the multicolinearity among the columns.
#if you vif >= 6 it means that column is not important for us , as it having high multicolinearity . we will remove those
```

```python
col_list = []
for x in df.columns:
    if df[x].dtype!='object' and x!='charges':
        col_list.append(x)
```

```python
col_list
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
dummy = df[col_list]
```

```python
dummy
```

```python
pd.DataFrame(dummy.values)
```

```python
vif_data = pd.DataFrame()
vif_data['Feature'] = dummy.columns
vif_data
vif_list = []

for i in range(len(dummy.columns)):
    vif = variance_inflation_factor(dummy.values,i)
    vif_list.append(vif)
vif_data['VIF'] = vif_list
```

```python
vif_data
```
